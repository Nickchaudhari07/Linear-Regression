```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
Mileage = pd.read_csv('/content/Vehicle_Data.csv')
Mileage.head()

```

```python
Mileage.info()
```

```python
Mileage.isnull().sum()
Mileage= Mileage.fillna(Mileage['horsepower'].mean())
pd.options.display.max_columns=None
pd.options.display.max_rows=None
print(Mileage)


```

```python
sn.heatmap(data= Mileage.corr(),annot= True, linewidth= 0.5, cmap= 'Blues', cbar= True, vmin= -1, vmax= 1 )
```

```python
#Spliting the data and Target
X= Mileage.drop(['name','Kilometer_per_liter','origin','year'], axis = 1)
Y= Mileage['Kilometer_per_liter']
sn.pairplot(
    Mileage,
    y_vars=["Kilometer_per_liter"],
    x_vars=["cylinders", "displacement","horsepower","weight","acceleration"])
```

```python
print(X)
```

```python
print(Y)
```

```python
#Spliting Training and Test Data
X_train,X_test,Y_train,Y_test= train_test_split(X, Y, test_size= 0.1, random_state= 2)
```

```python
#Loading the linear Regression model
lin_reg_model= LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

```python
Test_data_prediction = lin_reg_model.predict(X_test)
```

```python
#R squared Error
error_score= metrics.r2_score(Y_test, Test_data_prediction)
print('R Squared Error:',error_score)
```

```python
#Visualization
sn.regplot(x=Y_test, y=Test_data_prediction)
plt.xlabel('Actual Mileage')
plt.ylabel('Predicted Mileage')
plt.title('Actual Vs Predicted Mileage')
```

```python
#Sum of Squared Error
SSE= metrics.mean_squared_error(Y_test, Test_data_prediction)
print('SSE:',SSE)
```

```python
#Mean Absolute Error
MAE= metrics.mean_absolute_error(Y_test, Test_data_prediction)
print('MAE:',MAE)
```

```python
#Mean Squared Log Error
MSLE= metrics.mean_squared_log_error(Y_test, Test_data_prediction)
print('MSLE:',MSLE)
```
